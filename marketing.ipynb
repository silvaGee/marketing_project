```python
df.drop("ID", axis=1, inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={' Income ':'Income', 'Response':'AcceptedCmp6'}, inplace=True)
df.head()
```

```python
df['Income'] = df['Income'].str.replace('[$,]', '', regex=True).astype(float)
df
```

```python
missing_income = df[df["Income"].isna()]
missing_income
```

```python
mean_phd_married =(df["Education"]=="PhD") & (df["Marital_Status"]=="Married") & (df["Income"].notna())
df[mean_phd_married]
```

```python
df[mean_phd_married].info()
```

```python
df[mean_phd_married].describe()
```

```python
df[mean_phd_married]["Income"].mean()
```

```python
# for index, row_series in df.iterrows() - to iterate over rows
#missing_income = variable for na values in income column
#go through missing income rows and impute the values based on the mean of the same edu level and marital status
#mean_phd_married =(df["Education"]=="PhD") & (df["Marital_Status"]=="Married") & (df["Income"].notna())

for index, row_series in missing_income.iterrows():
    mean_income = (df["Education"]==row_series["Education"]) & (df["Marital_Status"]==row_series["Marital_Status"]) & (df["Income"].notna())
    df.loc[index, "Income"] = df[mean_income]["Income"].mean()
```

```python
df.info()
```

```python
#show labels for Marital Status
df["Marital_Status"].value_counts()
```

```python
#Dropping row with Alone, YOLO, and Absurd marital status because they are unclear and data is less than 5% of all data
df.drop(df.loc[df["Marital_Status"] == "YOLO"].index, inplace=True)
df["Marital_Status"].value_counts()
```

```python
df.drop(df.loc[df["Marital_Status"] == "Alone"].index, inplace=True)
df["Marital_Status"].value_counts()
```

```python
df.drop(df.loc[df["Marital_Status"] == "Absurd"].index, inplace=True)
df["Marital_Status"].value_counts()
```

```python
# data = {'Size': ['Small', 'Medium', 'Large', 'Medium', 'Small']}
# df = pd.DataFrame(data)

# # Define the ordered categories
# ordered_categories = pd.CategoricalDtype(categories=['Small', 'Medium', 'Large'], ordered=True)

# # Convert the column to ordered categorical
# df['Size_Ordered'] = df['Size'].astype(ordered_categories)

# # Access the category codes (numerical representation)
# df['Encoded_Size'] = df['Size_Ordered'].cat.codes
df["Marital_Status"] = df["Marital_Status"].astype("category")
df.info()
```

```python
df['Education'].value_counts()
```

```python
ordered_categories = pd.CategoricalDtype(categories=['Basic', 'Graduation', '2n Cycle', 'Master', 'PhD'], ordered=True)
df['Education'] = df['Education'].astype(ordered_categories)
df['EducationEncoded'] = df['Education'].cat.codes
df.info()
```

```python
df.rename(columns={'Dt_Customer':'DateEnrolled', 'Recency':'DaysSincePurchase', 'Marital_Status': 'MaritalStatus', 'Year_Birth':'BirthYear'}, inplace=True)
df.info()
```

```python
df.drop(df.loc[df["Country"] == "ME"].index, inplace=True)
df["Country"].value_counts()
```

```python
df["Country"] = df["Country"].astype("category")
df.info()
```

```python
df.head()
```

```python
df['DateEnrolled'] = pd.to_datetime(df['DateEnrolled'], infer_datetime_format=True)
df.info()
```

```python
df.head()
```

```python
from datetime import datetime
current_year = datetime.now().year
df['Age'] = current_year - df['BirthYear']
df.head()
```

```python
df.drop('BirthYear', axis=1, inplace=True)
df.head()
```

```python
#Total # of children
df['TotalNumChildren'] = df['Kidhome'] + df['Teenhome']
df.head()
```

```python
#Total Spent
df['TotalSpent'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df.head()
```

```python
df['TotalPurchases'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']
df.head()
```
