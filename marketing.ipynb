```python
# for index, row_series in df.iterrows() - to iterate over rows
#missing_income = variable for na values in income column
#go through missing income rows and impute the values based on the mean of the same edu level and marital status
#mean_phd_married =(df["Education"]=="PhD") & (df["Marital_Status"]=="Married") & (df["Income"].notna())

for index, row_series in missing_income.iterrows():
    mean_income = (df["Education"]==row_series["Education"]) & (df["Marital_Status"]==row_series["Marital_Status"]) & (df["Income"].notna())
    df.loc[index, "Income"] = df[mean_income]["Income"].mean()
```

```python
#Find out if people with children do more online shopping vs other forms
#Find out if other forms of purchases are canabalizing store purchases
matrix_features = ['TotalNumChildren', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

matrix = df[matrix_features].corr()
plt.figure(figsize=(10, 10))
sns.heatmap(matrix, annot=True)
plt.title('Correlation Matrix of Chilren at Home vs. Shopping Preference')
plt.show()
print(matrix)
```

According to the heat map, people with children at home are slightly more likely to make web purchases vs other mediums like store/catalog purchases.  If we had time series data that showed the history of each type of purchase, we could say with more certainty if there was canabalization.  However, with the current data, other forms of purchases such as catalog and web purchases do not seem to be canabalizing store purchases since they are positively correlated.  If anything, this tells us that customers that are making purchases are using multiple mediums to make purchases.

```python
from sklearn.linear_model import LinearRegression


# Define independent and dependent variables
X = df[['NumWebPurchases', 'NumCatalogPurchases']]
y = df['NumStorePurchases']

# Fit the regression model
model = LinearRegression()
model.fit(X, y)

# Print coefficients and intercept
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Optionally, display which coefficient corresponds to which variable
for col, coef in zip(X.columns, model.coef_):
    print(f"{col}: {coef}")
```

```python
print(X.shape)
```

```python
print(y.shape)
```

```python
import matplotlib.pyplot as plt

# Predict store purchases using the model
y_pred = model.predict(X)

# Scatter plot: Actual vs Predicted
plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred, alpha=0.6)
plt.xlabel('Actual Store Purchases')
plt.ylabel('Predicted Store Purchases')
plt.title('Actual vs Predicted Store Purchases')
# Diagonal line
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.show()

# Residual plot
residuals = y - y_pred
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, alpha=0.6)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted Store Purchases')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Store Purchases')
plt.show()
```

```python
print(y_pred.shape)
```

```python
# Get feature names and corresponding coefficients
features = list(X.columns)
coefficients = model.coef_
intercept = model.intercept_

# Combine intercept and coefficients for plotting
labels = ['Intercept'] + features
values = [intercept] + list(coefficients)

# Create bar plot
plt.figure(figsize=(8, 5))
plt.bar(labels, values, color='skyblue')
plt.title('Linear Regression Coefficients and Intercept')
plt.ylabel('Value')
plt.show()
```
